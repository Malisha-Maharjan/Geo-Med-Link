```python
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
train, eva = train_test_split(data, test_size=0.2)
```

```python
import simpletransformers

from simpletransformers.classification import ClassificationModel


```

```python
model = ClassificationModel('bert', 'bert-base-cased', num_labels=2, args={'reprocess_input_data': True, 'overwrite_output_dir': True}, use_cuda=False)
```

```python
model.train_model(train)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(eva)
```

```python
result

```

```python
lst = []
for arr in model_outputs:
    lst.append(np.argmax(arr))
```

```python
true = eva['is_spam'].tolist()
predicted = lst
```

```python
import sklearn
mat = sklearn.metrics.confusion_matrix(true, predicted)
mat
```

```python
sklearn.metrics.accuracy_score(true, predicted)
```

```python
import matplotlib.pyplot as plt
df_cm = pd.DataFrame(mat, range(2), range(2))
sns.heatmap(df_cm, annot=True)
plt.show()
```

```python
def get_result(statement):
    result = model.predict([statement])
    pos = np.where(result[1][0] == np.amax(result[1][0]))
    pos = int(pos[0])
    sentiment_dict = {0:'not spam', 1:'spam'}
    print(sentiment_dict[pos])
    return
```

```python
from transformers import BertTokenizer, BertModel
import matplotlib.pyplot as plt
import seaborn as sns

# Load a pre-trained BERT model and its tokenizer
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# Input text
text = "Replace this with your text for analysis."

# Tokenize the text and obtain attention weights
inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
outputs = model(**inputs)
attention = outputs.attentions

# Plotting the attention weights for a specific layer and attention head
layer = 0  # Choose the layer you want to visualize
head = 0   # Choose the attention head within that layer

plt.figure(figsize=(10, 8))
sns.heatmap(attention[layer][0][head].detach().numpy(), cmap='viridis', linewidths=0.5)
plt.title(f'Attention Weights - Layer {layer}, Head {head}')
plt.xlabel('Token IDs')
plt.ylabel('Token IDs')
plt.show()
```
